# Website Traffic Forecasting

The dataset contains information about daily traffic data from June 2021 to June 2022

## importing the necessary Python libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
```

### Reading CSV

```python
data = pd.read_csv("traffic.csv")

print(data.head())
```

### converting the Date into Datetime 

```python
data["Date"] = pd.to_datetime(data["Date"],format="%d/%m/%Y")
print(data.info())
```

### daily traffic of the website

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.plot(data["Date"], data["Views"])
plt.title("Daily Traffic")
plt.show()
```

### showing if  dataset is stationary or seasonal:

```python
result = seasonal_decompose(data["Views"], model='multiplicative', period = 30)
fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(15, 10)
```

#### the data is not stationary, 
the value of d is 1. 
using  the autocorrelation and partial autocorrelation plots to determine p and q:

```python
pd.plotting.autocorrelation_plot(data["Views"])
```

```python
plot_pacf(data["Views"], lags = 100)
```

### SARIMA Model training

```python
p, d, q = 5, 1, 2
model=sm.tsa.statespace.SARIMAX(data['Views'],
                                order=(p, d, q),
                                seasonal_order=(p, d, q, 12))
model=model.fit()
print(model.summary())
```

### 50-day Website traffic forecasting

```python
predictions = model.predict(len(data), len(data)+50)
print(predictions)
```

###  predictions plot

```python
data["Views"].plot(legend=True, label="Training Data", 
                   figsize=(15, 10))
predictions.plot(legend=True, label="Predictions")
```
